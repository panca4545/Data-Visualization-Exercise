```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import folium

pd.set_option('display.max_columns',500)
```

```python
df=pd.read_csv('melb_data.csv')
```

## Nomor 1
a) Buatlah Histogram untuk menampilkan distribusi data harga rumah (kolom “Price”)!

b) Tambahkan judul “Distribusi Harga Rumah” berukuran 15!

c) Tidak ada ketentuan pilihan warna, ukuran, dan library yang digunakan.

```python
plt.hist(df['Price'])
plt.title('Distribusi Harga Rumah', size=15)
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()

# masi bingung, kenapa itu 0-8 sedangkan kolom price kan ratusan ribu

# Frequency emang keyword dari sananya ya? 
#kok kalo saya ganti rooms hasilnya sama aja?
```

```python
plt.style.use('seaborn')
plt.figure(figsize=(8,6))
plt.hist(df['Price'],10,color='red')
plt.title('Distribusi Harga Rumah',size=15)
plt.xlabel('Price',size=10)
plt.ylabel('Price',size=10)
plt.grid(True)
plt.legend(['Distribusi Harga Rumah'],loc=0)
plt.savefig('Distribusi Harga Rumah.png')
plt.show()


```

## No 3
a) Buatlah Bar Plot (diagram batang) untuk membandingkan nilai median harga 
rumah (kolom “Price”) di tiap tipe rumah (kolom “Type”)!

b) Tambahkan judul ‘Perbandingan Nilai Median Harga Rumah’ berukuran 15!

c) Tidak ada ketentuan pilihan warna, ukuran, dan library yang digunakan.

```python
from numpy import median
sns.barplot(data=df, x='Type',y='Price',estimator=median)
plt.title('Perbandingan Nilai Median Harga Rumah', size=15)
plt.show()

#Masih bingung cara baca datanya
```

## NO 5
a) Konversi kolom “Date” menjadi tipe data datetime64. Buatlah kolom baru 
bernama ‘Month’ hasil ekstraksi data bulan dari kolom “Date”!

b) Buatlah Line Plot (diagram garis) untuk menampilkan tren perkembangan 
harga rumah (kolom “Price”) tiap bulan!

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.info()
```

```python
df['Month'] = pd.DatetimeIndex(df['Date']).month
df.head()
```

```python
sns.lineplot(data=df, x='Month', y='Price', hue='Type')
plt.title('Tren Harga Rumah per Bulan berdasarkan Tipe Rumah', size=15)
plt.show()

```
